##source_crm

### Lendo os dados de arquivos .csv 
#### (sales_details)

```python
df_sales =(
  spark.read.option("header", "True")
  .option("inferSchema", "True")
  .csv("/Volumes/workspace/bronze/source_systems/source_crm/sales_details.csv")
)
df_sales.display()
```

### Escrevendo dados na camada Bronze
#### (crm_sales_details)

```python
df_sales.write.mode("overwrite").saveAsTable("bronze.crm_sales_details")
```

```python
%sql
-- Consultando os dados na tabela criada
select * from bronze.crm_sales_details
```
